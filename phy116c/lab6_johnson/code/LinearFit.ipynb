```python
x = numpy.array([1, 5, 15, 30])
y = numpy.array([12,20,40, 70])
plot(x,y,"o")
from scipy.optimize import curve_fit
def line(x, a, b):
    return a * x + b
popt, pcov = curve_fit(line, x, y)
print popt
xfine = np.linspace(0., 30., 100)
plt.plot(xfine, line(xfine, popt[0], popt[1]), 'r-')
```

```python
FKT = 1.64E-20
G = 4000
BW = 20E3
def vs(r):
    return 1000*(FKT * r * G * G * BW)**(0.5)

r = numpy.array([30.0, 60.0, 100.0, 500.0, 1000.0])
v = vs(r*1000)

print v

plot(r,v**2,"o")
popt, pcov = curve_fit(line, r, v**2)
print popt
xfine = np.linspace(0., 1000., 100)
plt.plot(xfine, line(xfine, popt[0], popt[1]), 'r-')
show();

vfine = sqrt(abs(line(xfine, popt[0], popt[1])))
plot(r,v,"o")
plot(xfine,vfine,"-")
show();
```
