```python
max_samples = 1500
x = np.arange(max_samples)
y = sin(15*2*pi*x/max_samples)
plot(x,y)
```

```python
from scipy import signal
fs=76.900 # sample frequency in Hz
f, psd = signal.periodogram(y, fs)
f = f[0:100]
psd = psd[0:100]
plt.semilogy(f, psd)
#plt.plot(f, psd)
plt.xlabel('frequency [kHz]')
plt.ylabel('PSD [V**2/kHz]')
plt.show()
```
